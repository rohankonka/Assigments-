```python
# Import Libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import  AgglomerativeClustering
from sklearn.preprocessing import  normalize
from sklearn.cluster import KMeans
from sklearn import preprocessing
import seaborn as sns
```

```python

airline=pd.read_csv('Downloads/EastWestAirlines (4).csv')
airline
```

```python
airline.info()
```

```python
airline2=airline.drop(['ID#'],axis=1)
airline2
```

## H-Clustering

```python

airline2_norm=pd.DataFrame(normalize(airline2),columns=airline2.columns)
airline2_norm
```

```python

plt.figure(figsize=(10,7))
dendograms=sch.dendrogram(sch.linkage(airline2_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(airline2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python

airline2['clustersid']=hclusters.labels_
airline2
```

```python
airline2.groupby('clustersid').agg(['mean']).reset_index()
```

```python

plt.figure(figsize=(10, 7))  
plt.scatter(airline2['clustersid'],airline2['Balance'], c=hclusters.labels_) 
```

## KMeans Clustering

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python

airline=pd.read_csv('Downloads/EastWestAirlines (4).csv')
airline
```

```python
airline2=airline.drop(['ID#'],axis=1)
airline2
```

```python
airline2.info()
```

```python

airline2_norm=StandardScaler().fit_transform(airline2)
```

```python

wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(airline2_norm)
    wcss.append(kmeans.inertia_)
```

```python

plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters4=KMeans(4,random_state=30).fit(airline2_norm)
clusters4
```

```python
clusters4.labels_
```

```python

airline4=airline2.copy()
airline4['clusters4id']=clusters4.labels_
airline4
```

```python

clusters4.cluster_centers_
```

```python

airline4.groupby('clusters4id').agg(['mean']).reset_index()
```

```python

plt.figure(figsize=(10, 7))  
plt.scatter(airline4['clusters4id'],airline4['Balance'], c=clusters4.labels_)
```

```python
clusters5=KMeans(5,random_state=30).fit(airline2_norm)
clusters5
```

```python
clusters5.labels_
```

```python

airline5=airline2.copy()
airline5['clusters5id']=clusters5.labels_
airline5
```

```python

clusters5.cluster_centers_
```

```python

airline5.groupby('clusters5id').agg(['mean']).reset_index()
```

```python

plt.figure(figsize=(10, 7))  
plt.scatter(airline5['clusters5id'],airline5['Balance'], c=clusters5.labels_)
```

## DBSCAN

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Downloads/EastWestAirlines (4).csv')
df.head()
```

```python
df.drop(['ID#','Award?','cc1_miles','cc2_miles','cc3_miles'],axis=1,inplace=True)
```

```python
array=df.values
array
```

```python
stscaler=StandardScaler().fit(array)
stscaler
```

```python
x=stscaler.transform(array)
```

```python
dbscan=DBSCAN(eps=.8,min_samples=7)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['clusters'])
```

```python
df1=pd.concat([df,cl],axis=1)
df1
```

```python
dbscan.core_sample_indices_
```

```python
df1.groupby('clusters').count()
```

```python
df1.groupby('clusters').agg(['mean']).reset_index()
```
